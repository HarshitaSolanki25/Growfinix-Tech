```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='survived',data=df)
plt.title('Survival Count')
plt.savefig('Survival_Distribution.png')
plt.show()
```

### Survival Distribution
The data Set shows that the number of passengers who did not survive was higher that the number of passengers who survived. This indicates that the survival rate on the Titanic was less than 50%.


```python
sns.countplot(x='sex',hue='survived',data = df)
plt.title('Survival by gender')
plt.savefig('survival_by_gender')
plt.show()
```

### Survival by Gender

Female passengers had a significantly higher survival rate compared to male passengers. This suggests that women were given the evacuation process.



```python
sns.countplot(x='class',hue='survived',data=df)
plt.title('Survival by passenger class')
plt.savefig('survival_by_class.png')
plt.show()
```

### Survival by Passenger Cass

Passenger class had a significant impact on survival. First-class passengers had the highest survival rate, while third-class paseengers had the lowest survival rate. This suggests that passengers in higher classes had better access to lifeboats and evacuation facilities.

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.savefig('age_distribution.png')
plt.show()
```

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

```python
sns.barplot(x='family_size', y='survived', data=df)
plt.title('Survival Rate by Family Size')
plt.savefig('family_size_survival.png')
plt.show()
```

### Survival Rate by Family Size
Passengers travelling with small families had a higher survival rate compared to those travelling alone or in very large families.

```python
sns.barplot(x='survived', y='fare', data=df)
plt.title('Average Fare by Survival')
plt.savefig('fare_vs_survival.png')
plt.show()
```

### Fare vs Survival
Passengers who paid higher ticket fares generally had a higher survival rate. This suggests that passengers in higher classes had better chances of Survival.

```python
df.groupby('sex')['survived'].mean()
```

```python
df.groupby('class', observed=False)['survived'].mean()
```

```python
pd.pivot_table(
    df,
    values='survived',
    index='sex',
    columns='class',
    observed = 'false',
    aggfunc='mean'
)
```

### GroupBy and Pivot Table Insights

GroupBy and Pivot Table analysis show that female passengers and first-class passengers had significantly higher survival rates.

### Final Conclusion



1. Female passengers had a much higher survival rate than male passengers.
2. First-class passengers were more likely to survive than second and third-class passengers.
3. Most passengers were between 20 and 40 years old.
4. Higher ticket fares were associated with better survival chances.
5. Passengers traveling with small families had better survival rates.
6. Gender and passenger class were the strongest factors affecting survival.
